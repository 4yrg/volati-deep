## Fetch Dataset from yfinance

```python
import yfinance as yf

# List of stock trickers
trickers = ['AAPL', 'NVDA', 'MSFT', 'JPM', 'V', 'NKE', 'MCD', 'JNJ', "PFE"]

# Define the data range
start_date = '2015-01-01'
end_date = '2025-09-15'

# Download the dataset from yfinance
stock_data = yf.download(trickers, start=start_date, end=end_date)

# Save the DataFrame to the specific path
file_path = "/content/drive/MyDrive/SLIIT/Y4S1-SE4050-DL/DL-ASSIGNMENT/volati-deep/data/raw/raw_stock_prices.csv"
stock_data.to_csv(file_path)

print(f"Data successfully saved to {file_path}")
```
